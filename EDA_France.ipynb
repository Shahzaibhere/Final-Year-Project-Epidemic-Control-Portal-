<h2 id="load_dataset">1. Loading the clean dataset</h2>

```python
import pandas as pd

df=pd.read_csv("Meteorological_dataset_global.csv",  parse_dates=['Date'], index_col=['Date'])
df.tail(50)
```

<h2 id="load_dataset">2. Exploratory Data Analysis </h2>

<h1 align=center><font size="3"> Hypothesis </font></h1>

> *  The avg temp is highly positive correlated with number of cases
> *  The humidity is highly positive correlated with number of cases
> *  The wind speed is highly negative correlated with number of cases
> *  The sun hour is highly positive correlated with number of cases



```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

<h1 align=center><font size="3"> French Polynesia, France </font></h1>

```python
df_F=df.loc[df['State']=='French Polynesia']
df_F.head(10)
```

```python
def line_plot(x,y,xlabel,ylabel,title):
    fig,ax=plt.subplots()
    ax.plot(x,y,color='r')
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    plt.show()
    
```

```python
line_plot(df_F.index,df_F["Confirmed"],"Date","Cases","Confirmed cases vs Date")

```

```python
line_plot(df_F.index,df_F["Temp / C"],"Date","Average Temperature","Average Temperature vs Date")

```

```python
df_F[["Confirmed","Temp / C"]].corr()
```

```python
sns.regplot(x="Confirmed", y="Temp / C", data=df_F)
plt.ylim(0,)
```

```python
df_F[["Confirmed","Humidity"]].corr()
```

```python
sns.regplot(x="Confirmed", y="Humidity", data=df_F)
plt.ylim(0,)
```

```python
df_F[["Confirmed","Wind Speed / KMPH"]].corr()
```

```python
sns.regplot(x="Confirmed", y="Wind Speed / KMPH", data=df_F)
plt.ylim(0,)
```

```python
df_F[["Confirmed","Sun Hour"]].corr()
```

```python
sns.regplot(x="Confirmed", y="Sun Hour", data=df_F)
plt.ylim(0,)
```
