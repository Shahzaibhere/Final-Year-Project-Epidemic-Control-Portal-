<h2 id="load_dataset">1. Loading the clean dataset</h2>

```python
import pandas as pd

df=pd.read_csv("Meteorological_dataset_global.csv",  parse_dates=['Date'], index_col=['Date'])
df.head(50)
```

<h2 id="load_dataset">2. Exploratory Data Analysis </h2>

<h1 align=center><font size="3"> Hypothesis </font></h1>

> *  The avg temp is highly positive correlated with number of cases
> *  The humidity is highly positive correlated with number of cases
> *  The wind speed is highly negative correlated with number of cases
> *  The sun hour is highly positive correlated with number of cases



```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

<h1 align=center><font size="3"> Hubei, China </font></h1>

```python
df_H=df.loc[df['State']=='Hubei']
df_H.head(10)
```

```python
def line_plot(x,y,xlabel,ylabel,title):
    fig,ax=plt.subplots()
    ax.plot(x,y,color='r')
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    plt.show()
    
```

```python
line_plot(df_H.index,df_H["Confirmed"],"Date","Cases","Confirmed cases vs Date")

```

```python
line_plot(df_H.index,df_H["Temp / C"],"Date","Average Temperature","Average Temperature vs Date")

```

```python
df_H[["Confirmed","Temp / C"]].corr()
```

```python
sns.regplot(x="Confirmed", y="Temp / C", data=df_H)
plt.ylim(0,)
```

```python
df_H[["Confirmed","Humidity"]].corr()
```

```python
sns.regplot(x="Confirmed", y="Humidity", data=df_H)
plt.ylim(0,)
```

```python
df_H[["Confirmed","Wind Speed / KMPH"]].corr()
```

```python
sns.regplot(x="Confirmed", y="Wind Speed / KMPH", data=df_H)
plt.ylim(0,)
```

```python
df_H[["Confirmed","Sun Hour"]].corr()
```

```python
sns.regplot(x="Confirmed", y="Sun Hour", data=df_H)
plt.ylim(0,)
```

<h2 id="load_dataset">3. Conclusions </h2>

> *  The average temperature in the four most affected Chinese provinces is only just mildly positively correlated with the number of confirmed cases.
> * The humidity percentage in these provinces was infact mildly negatively correlated with the confirmed cases.
> * Contrary to my hypothesis wind speed proved to be mildly positively correlated.
> * The sun hour metric didn't yield a consistent analysis as its results differed within the provinces. Hence, it is not a good feature for our analysis.
